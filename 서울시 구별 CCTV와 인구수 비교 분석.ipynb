```python
fp1 = np.polyfit(data['인구수'], data['소계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize= (6, 6))
plt.scatter(data['인구수'],data['소계'], s = 50)
plt.plot(fx, f1(fx), ls = 'dashed', lw =3, color = 'r')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
data['오차'] = np.abs(data['소계']- f1(data['인구수']))

df_sort = data.sort_values(by = '오차', ascending = False)
df_sort.head(10)
```

```python
plt.figure(figsize= (20, 8))
plt.scatter(data['인구수'],
            data['소계'], 
            c = data['오차'], 
            s = 50)
plt.plot(fx, 
         f1(fx), 
         ls = 'dashed', 
         lw =3, 
         color = 'r')

for n in range(10):
    plt.text(df_sort['인구수'][n]*1.015, 
             df_sort['소계'][n]*0.98, 
             df_sort.index[n],
             fontsize = 12)
    
plt.xlabel('인구수')
plt.ylabel('인구당비율')

plt.colorbar()
plt.grid()
plt.show()
```
